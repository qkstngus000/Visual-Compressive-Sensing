```python
#import image information
image_nm = 'city_part2.png'
img = Image.open('image/' + image_nm)
img = ImageOps.grayscale(img)
img_arr = np.asarray(img)

plt.figure(dpi=300, figsize=(3, 3))
plt.imshow(img, 'gray')
# plt.title("Original Image")
plt.axis('off')
dim = img_arr.shape
n, m = dim
num_cell = 500
plot = np.zeros(n*m)
rand = np.arange(0, n*m)
rand = np.random.permutation(rand)[:num_cell]
plot[rand] = 1
plot = plot.reshape(n, m)

x, y = np.where(plot == 1)

plt.plot(x, y, 'r.')
plt.show()
```
